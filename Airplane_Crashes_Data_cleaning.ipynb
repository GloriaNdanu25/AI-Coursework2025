```python
import zipfile
import os

with zipfile.ZipFile("archive.zip", "r") as zip_ref:
    zip_ref.extractall("data")
    import os
os.listdir("data")
```

```python
import pandas as pd

# Load the dataset from the extracted folder
df = pd.read_csv("data/Airplane_Crashes_and_Fatalities_Since_1908.csv")

# Check the shape and preview a few rows
print("Dataset loaded successfully!")
print("Shape:", df.shape)
df.head()
```

```python
# Q2
n_rows, n_cols = df.shape
print("Number of rows:", n_rows)
print("Number of columns:", n_cols)
```

```python
# Q3
df.tail(75)
```

```python
# Q8
# Create a smaller dataframe with the needed columns
fatality_locations = df[['Date', 'Location', 'Aboard', 'Fatalities']].copy()

# Split 'Location' into 'Region' and 'State_Country'
split_cols = fatality_locations['Location'].astype(str).str.split(',', n=1, expand=True)
fatality_locations['Region'] = split_cols[0].str.strip()
fatality_locations['State_Country'] = split_cols[1].str.strip()

# Show sample results
fatality_locations[['Date', 'Location', 'Region', 'State_Country']].head(10)
```

```python
fatality_locations.to_csv("fatality_locations_cleaned.csv", index=False)

from google.colab import files
files.download("fatality_locations_cleaned.csv")
```
